Dane o pasażerach Titanica





Zbiór danych zawiera informacje o pasażerach RMS Titanic, który zatonął 15 kwietnia 1912 roku po zderzeniu z górą
lodową. Dane obejmują takie atrybuty jak klasa podróży, wiek, płeć, liczba rodzeństwa/małżonków na pokładzie,
liczba rodziców/dzieci na pokładzie, cena biletu oraz miejsce zaokrętowania.

Zbiór zawiera także informację o tym, czy pasażer przeżył katastrofę.

Titanic przewoził ponad 2,200 osób, z czego ponad 1,500 zginęło, co czyni tę katastrofę jedną z najbardziej
tragicznych w historii morskiej.

Kolumny:

* **pclass** - Klasa biletu
* **survived** - Czy pasażer przeżył katastrofę
* **name** - Imię i nazwisko pasażera
* **sex** - Płeć pasażera
* **age** - Wiek pasażera
* **sibsp** - Liczba rodzeństwa/małżonków na pokładzie
* **parch** - Liczba rodziców/dzieci na pokładzie
* **ticket** - Numer biletu
* **fare** - Cena biletu
* **cabin** - Numer kabiny
* **embarked** - Port, w którym pasażer wszedł na pokład (C = Cherbourg, Q = Queenstown, S = Southampton)
* **boat** - Numer łodzi ratunkowej
* **body** - Numer ciała (jeśli pasażer nie przeżył i ciało zostało odnalezione)
* **home.dest** - Miejsce docelowe

```python
import pandas as pd
```

```python
# Wez liczbe z konca string 
# i zamien na int
def get_number_from_string(s):
    print(s)
    s=s.strip()
    if s == "":
       return 0
    if s is None:
       return 0
    sp=(s.split()[-1])
    if len(sp)==1:
        if sp.isdigit():
            return 1
        else:
            return 2
    elif len(sp)==2:
        return 3
    else:
        return 4
```

```python
df = pd.read_csv('26__titanic.csv', sep=",", converters= {"ticket": str},)
df["ticket"]=df["ticket"].apply(get_number_from_string)
```

```python
df.info()
```

```python
df['age'].fillna(df['age'].median(), inplace=True)
df
```

```python
# Usunięcie wierszy, w których brakuje kluczowych informacji (np. brak klasy, informacji o przeżyciu lub płci)
df_clean = df.dropna(subset=['pclass', 'survived', 'sex'])

# Uzupełnienie brakujących wartości w kolumnie 'fare' medianą
df_clean['fare'].fillna(df_clean['fare'].median(), inplace=True)

# Uzupełnienie brakujących wartości w kolumnie 'embarked' najczęstszą wartością (modą)
df_clean['embarked'].fillna(df_clean['embarked'].mode()[0], inplace=True)

# Wyświetlenie podstawowych informacji o oczyszczonym zbiorze
df_clean.info()
```

```python
lista_indeksow=[0,3,4,7]
#wybierz kolumny o indeksach 0, 3, 4 i 7
wybrane_kolumny = df_clean.iloc[:, lista_indeksow]

wybrane_kolumny["sex"]=wybrane_kolumny["sex"].apply(lambda x: 1 if "male" == x else 0)
wybrane_kolumny
```

```python
wybrane_kolumny.info()
```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Obliczanie macierzy korelacji
corr_matrix = wybrane_kolumny.corr()

# Tworzenie wykresu korelacji
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Macierz korelacji')
plt.show()
```

```python

# tworzymy zmienną która zlicza ilość i wartości rekordów w tabeli 'sex'
sex_counts = df['sex'].value_counts()

# tworzymy wykres
plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=['Male', 'Female'], autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Sex')
plt.axis('equal')
plt.show()
```

```python

age_data = df['age']

# tworzymy nową zmienną dla wartości z `age_data` w przedziale powyżej 16 oraz mniejszym/równym 16
age_counts = {
    '≤ 18': (age_data <= 18).sum(),
    '> 18': (age_data > 18).sum()
}

fig, ax = plt.subplots()
ax.pie(age_counts.values(), labels=age_counts.keys(), autopct='%1.1f%%', startangle=90)
ax.axis('equal') 
plt.show()
```

```python

# tworzymy zmienną, ktora zawiera sumy unikalnych wartości z kolumny 'survived'
survived_counts = df['survived'].value_counts()

# zmieniamy wyświetlane na wykresie nazwy dla wartości 1.0 i 0.0 
survived_counts.index = ['Survived' 
                         if i == 1.0 
                         else 'Dead' 
                         for i in survived_counts.index]

# tworzymy wykres kołowy pie chart
# 'labels' definiuje etykiety dla każdego segmentu wykresu kołowego, 'autopct' definiuje format procentowy, zaś 'startangle' obraca wykres o 90 stopni
plt.figure(figsize=(6, 6))
plt.pie(survived_counts, labels=survived_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Survival Distribution')

plt.show()
```

```python
# tworzymy zmienną, która grupuje dane po tabelach 'sex' i 'survived' 
survival_counts = df.groupby(['sex', 'survived']).size().unstack()

# tworzymy nowy wykres o wymiarach 1x2 z rozmiarem figury (12,6). Zmienne 'fig' i 'ax' są używane do manipulowania wykresem i jego podwykresami
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# zmieniamy wyświetlane na wykresie nazwy dla wartości 1.0 i 0.0 
survival_counts.columns = survival_counts.columns.map({0.0: 'Dead', 1.0: 'Survived'})

# wykres kołowy dla mężczyzn
ax[0].pie(survival_counts.loc['male'], labels=survival_counts.columns, autopct='%1.1f%%', startangle=90)
ax[0].set_title('Survival Distribution for Males')

# wykres kołowy dla kobiet
ax[1].pie(survival_counts.loc['female'], labels=survival_counts.columns, autopct='%1.1f%%', startangle=90)
ax[1].set_title('Survival Distribution for Females')
plt.show()
```

```python
df = df.copy()
body_data = pd.to_numeric(df['body'], errors='coerce')  # zabezpieczenie

grouped_data_all = {
    'Survived': (body_data == 100).sum(),
    'Dead but no body found': (body_data > 99).sum(),
    'Dead and body found': (body_data < 99).sum()
}

grouped_data_dead = {
    'Dead but no body found': (body_data >= 99).sum(),
    'Dead and body found': (body_data < 99).sum()
}

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].pie(grouped_data_all.values(), autopct='%1.1f%%')
ax[0].set_title('Distribution of Body Values')
legend_0 = ax[0].legend(grouped_data_all.keys(), bbox_to_anchor=(0.5, -0.05), loc='upper center', title="Legend")

ax[1].pie(grouped_data_dead.values(), autopct='%1.1f%%')
ax[1].set_title('Distribution of Dead Victims')
legend_1 = ax[1].legend(grouped_data_dead.keys(), bbox_to_anchor=(0.5, -0.05), loc='upper center', title="Legend")

plt.tight_layout()
plt.show()
```
